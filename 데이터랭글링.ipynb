## 2-3 데이터프레임 탐색하기 loc iloc
*데이터프레임의 개별 데이터나 일부를 선택해야함. -> loc나 iloc 메서드를 사용해 하나 이상의 행이나 값을 선택 


```python
#라이브러리 임포트
import pandas as pd
#엑셀 파일 경로
file_path = 'C:/Users/smcuser/Desktop/[DRB]ver.2/test.xlsx'
#엑셀 파일 불러오기
#sheet_name  엑셀 파일 내에서 데이터를 읽어올 특정 시트 지정, 기본 값은 0이고 이는 첫 번째 시트를 가져온다는 뜻 혹은 직접 이름을 적어도 된다.
#header는 데이터프레임에서 열 이름으로 사용할 엑셀 시트의 행을 지정. 정수로 지정되며 기본값은 0이다. 이는 엑셀 시트의 첫 번째 행을 말한다.
df = pd.read_excel(file_path,sheet_name=0,header=1)
```

```python
#두번재 행 선택
df.iloc[1]
```

```python
#콜론(:) 을 사용하여 원하는 행의 슬라이스(slice)를 선택할 수 있다. 예를 들어 두 번째, 세 번재, 네 번재 행을 선택


df.iloc[1:4]
```

```python
df.iloc[:4]
```

```python
# 데이터프레임은 정수 인덱스를 굳이 가질 필요는 없습니다. 각 행이 고유하다면 어떤 값이라도 데이터프레임 인덱스로 만들 수 있습니다.

#인덱스 설정
dataframe = df.set_index(df['ID'])

#행을 확인

dataframe.loc[7777]
```

판다스 데이터프레임의 모든 행은 고유한 인덱스 값을 가집니다.
기본적으로 인덱스는 데이터프레임의 행 위치를 의미하는 정수이지만 필수는 아닙니다.
데이터프레임의 인덱스는 영문자나 숫자로 이루어진 고유한 문자열이거나 임의의 숫자일 수도 있습니다.
개별 행이나 행의 슬라이스를 선택하기 위해서 판다스는 두 개의 메서드를 제공합니다.

loc는 데이터 프레임의 인덱스 레이블(예를 들어 문자열)일 때 사용합니다.
iloc은 데이터프레임의 위치를 참조합니다. 예를들어 iloc[0]은 정수 혹은 문자열 인덱스에 상관없이 첫 번재 행을 반환합니다. 
데이터 정제 단계에서 자주 등장하기 때문에 loc 메서드와 iloc 메서드에 익숙해지는 것이 좋습니다.

## 2-8. 누락된 값 다루기 isnull과 notnull
데이터프레임에서 누락된 값을 선택하고 싶습니다.

```python
import pandas as pd

file_path = 'C:/Users/smcuser/Desktop/[DRB]ver.2/test.xlsx'

dataframe = pd.read_excel(file_path,sheet_name = 0, header = 1)

#누락된 값을 선택하고 두 개의 행을 출력합니다.

dataframe[dataframe['local recurrence site 2'].isnull()].head(2)
```

누락된 값은 데이터 랭글링에 아주 흔하게 등장하는 문제이지만 많은 사람이 누락된 값을 다루는 작업의 어려움을 과소평가합니다.
판다스 넘퍼아 NaN("Not A Number")을 사용하여 누락된 값을 사용하여 누락된 값을 표시합니다.
중요한 점은 판다스 자체적으로 NaN을 구현하지 않았다는 것입니다.
예를 들어 Apple이라는 문자열을 NaN으로 바꾸려고하면 에러가 발생합니다.
NaN을 사용하려면 넘파이 라이브러리를 임포트해야 합니다.


```python
#라이브러리를 임포트합니다.
import numpy as np

#NaN으로 값을 바꿉니다.
dataframe['나이'] = dataframe['나이'].replace('22',np.nan)
```
